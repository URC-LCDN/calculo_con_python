# Funciónes Pares e Impares

### Definición 1 _Función Par_

Una función $f:\mathbb{R}\to\mathbb{R}$ es **par**, si para cada $x$ en el dominio de $f$ se cumple que $$f(–x)=f(x).$$


Las funciones pares tienen simetría reflectiva a través del eje de las $y$.

### Ejemplo

El ejemplo canónico de una función par, es la función $f:\mathbb{R}\to\mathbb{R}$ definida como $f(x)=x^2$.

Por que se cumple que

$$f(x) = x^2 = x\cdot x = -x \cdot -x = (-x)^2 = f(-x)$$

En la figura 1, se observa el significado geométrico de una función par, donde su gráfica es simétrica respecto al eje $y$.

image.png

Figura 1 (J Stewart 2012, p. 18). Representación gráfica de una función par.

Otro ejemplo, es la funcion $cos(x)$.

### Definición 2 _Función Impar_

Una función $f:\mathbb{R}\to\mathbb{R}$ es **impar**, si para cada $x$ en el dominio de $f$ se cumple que $$f(–x) = -f(x).$$


Las funciones impares tienen simetría rotacional de $180^{\circ}$ con respecto del origen.    

El ejemplo canónico de una función par, es la función $f:\mathbb{R}\to\mathbb{R}$ definida como $f(x)=x^3$.

Por que se cumple que

$$f(-x) = (-x)^3 = -x\cdot -x \cdot -x = x^{2} \cdot -x = -(x^{3}) = -f(x)$$

En la figura 2, se observa el significado geométrico de una función impar.

Otro ejemplo, es la funcion $sen(x)$.

# Implementación en Python

#### Método Subs

Hacemos un parentesís para explicar el método [`subs`](https://docs.sympy.org/latest/tutorials/intro-tutorial/basic_operations.html) de `sympy`


Este método, se utiliza para evaluar en una expresión simbólica un valor numérico o simbólico, de la siguiente manera:

Consideremos la siguiente expresión simbólica

```python
import sympy as sp

# Definimos el simbolo
x, y = sp.symbols('x, y')
exp = x**2 + 2*x*y + y**2
```

Evaluación de una expresión con un valor numérico

```python
# x --> indica la variable que se va a evaluar en la expresión dada
# 1 --> indica el valor que tomara la variable indicada en el parametro anterior
exp_x = exp.subs(x, 1)
exp_x
```

Evaluación de varias variables en una misma expresión

Opción 1

Realizar la evaluación de forma anidada, de la forma siguiente:

```python
exp_x_y_opc_1 = exp.subs(x, 1).subs(y, 0)
exp_x_y_opc_1
```

Esta notación complica la sintaxís para cuando se realizan evaluaciones de muchas variables.  

Opción 2

Realizar la evaluación con una lista de tuplas, de la forma siguiente:

```python
exp_x_y_opc_2 = exp.subs([(x, 1), (y, 0)])
exp_x_y_opc_2
```

Esta notación es más facil de entender y automatizar cuando se realizan evaluaciones de muchas variables.

#### Implementación

Importamos el módulo `Sympy` para poder realizar cálculo simbólico.  

```python
import sympy as sp
```

En el bloque de código anterior la parte `as sp` indica un alías para el módulo `sympy`, esto es usual en Python para no escribir el nombre del modulo cada vez que sea necesario.

Siguiendo con el ejemplo de función par. Definimos la función $f(x)=x^{2}$.

```python
# Definimos el simbolo
x = sp.symbols('x')
# Definimos la funcion f(x) = x^{2}
f = x**2
```

Ahora calculemos los valores para $f(x)$ y $f(-x)$, y así verificar si la función $f(x)=x^{2}$ es par, como se vio en la sección anterior.   

```python
# f(x)
f_x = f.subs(x, x)
f_x
```

```python
#  f(-x)
f__x = f.subs(x, -x)
f__x
```

```python
f_x - f__x == 0
```

Lo anterior, indica que $f(x)$ es igual a $f(-x)$, por lo tanto $f(x)=x^{2}$ es una función par.

**Remark**
Cuando comparamos expresiones simbólicas $f_1$ y $f_2$, lo recomendable es restar las expresiones y comparar a cero ($f_1 - f_2 == 0$), que comparar directamente las expresiones ($f_1 == f_2); esto se debe a limitaciones del cómputo simbólico, es partircular en la forma en como se generan las expresiones simbólicas.
